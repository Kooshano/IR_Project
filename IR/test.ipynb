```python
from pathlib import Path
verbs_file_open = Path("verbs.dat").open(encoding="utf8")
verbs = verbs_file_open.readlines()
bons = {verb.split("#")[0] for verb in verbs}
verbe = set(
    [bon + "ه" for bon in bons]
    +
    ["ن" + bon + "ه" for bon in bons],
)
print(verbe)
```
